```python
## Reading the text file temperatures.txt

import string
import pandas as pd
with open('C:\\Users\\earora3\\Downloads\\temperatures.txt') as f:
    lines = f.readlines()
final=[]
for i in range(len(lines)):
    final.append(lines[i].strip('\n'))

final
```

```python
# Data cleaning function (converting string into int, removing punctuations)
def clean_data(data):
    temp=[]
    date=[]
    final_data=[]
    for i in data:
        text=i.strip()
        text = i.translate(str.maketrans('','',string.punctuation))
        text = [[int(s) for s in line.split(" ")] for line in text.split("\n")]
        final_data.append(text)
        
    return final_data
```

```python
df=clean_data(final)
df
```

```python
# Splitting data into part1 and part2
def split_data(df):
    part1=df[0:500]
    part2=df[500:]
    return part1,part2
```

```python
part1,part2=split_data(df)
```

```python
# Mapper function
# Input: part1 and part2 from split_data
# Output: List of dictionary pairs
def mapper(data):
    date = []
    temp = []
    result=[]
    final_data = []
    for i in data:
        for j in i:
            for k in range(len(j)):
                if k%2==0:
                    date.append(j[k])
                else:
                    temp.append(j[k])
    for i in range(len(date)):
        result.append({'date':date[i],'temp':temp[i]})
    return result   
   
```

```python
map_part1=mapper(part1)
map_part2=mapper(part2)

final_mapper=map_part1 + map_part2
print(final_mapper)
```

```python
# Function to sort output from mapper 1 and mapper 2
def sort(sorted_data):
    new_list = sorted(sorted_data, key=lambda d: d['date'])  
    return new_list
```

```python
# print(sort(final_mapper))
```

```python
# Storing the sort function output into a variable 
x=sort(final_mapper)
```

```python
# Splitting the date into year 
dataset=pd.DataFrame(x)
dataset['date'] = dataset['date'].astype(str)
dataset['Year'] = [x[:4] for x in dataset['date']]
```

```python
dataset.head()
```

```python
# partition of sorted key-value pairs
def partition(a):
    df1=a[a['Year']<='2015']
    df2=a[a['Year']>'2015']
    return df1,df2
```

```python
partition_1,partition_2=partition(dataset)
```

```python
#reducer function
def reducer(output):
    reducer_data=output.groupby(['Year'], as_index=False)['temp'].agg(list)
    max_list=[]
    for i in range(len(reducer_data)):
        max_list.append(max(reducer_data['temp'][i]))  
    reducer_data['Max_temp']=max_list
    return reducer_data
```

```python
# Calling output of partition function as input for reducer function
data1=reducer(partition_1)
```

```python
data2=reducer(partition_2)
```

```python
# Making changes according to the desired output asked in the assignment
data1=data1.drop('temp',axis=1)
data2=data2.drop('temp',axis=1)
```

```python
# Combining output of two partitions together
final_result=pd.concat([data1, data2], ignore_index=True)
```

```python
# Main function to wrap all the steps together
def main(x):
    df=clean_data(x)
    part1,part2=split_data(df)
    map_part1=mapper(part1)
    map_part2=mapper(part2)
    final_mapper=map_part1 + map_part2
    x=sort(final_mapper)
    partition_1,partition_2=partition(dataset)
    data1=reducer(partition_1)
    data2=reducer(partition_2)
    data1=data1.drop('temp',axis=1)
    data2=data2.drop('temp',axis=1)
    final_output=pd.concat([data1, data2], ignore_index=True)
    
    return final_output
    
    
```
